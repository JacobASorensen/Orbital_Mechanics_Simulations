```python
# Curtis Problem 6.8

import sympy
from sympy import *
import sys

# symbols

h_c_1   = Symbol( ' h_c_1'  ) # inner circle orbit altitude, km
h_c_2   = Symbol( ' h_c_2'  ) # inner circle orbit altitude, km
r_c_1   = Symbol( ' r_c_1'  ) # inner circle orbit altitude, km
r_c_2   = Symbol( ' r_c_2'  ) # inner circle orbit altitude, km
mu      = Symbol(  'mu'     ) # Earth geocentric gravitational constant, mu = 3.896 * 10**5
r_E     = Symbol( ' r_E')     # radius of Earth, r_E = 6378 km

# values

values = [
    (r_c_1, h_c_1 + r_E ),
    (r_c_2, h_c_2 + r_E ),
    (h_c_1, 300),
    (h_c_2, 3000),
    (mu, 3.986 * 10**5),
    (r_E, 6378),
]

# semimajor axis of transfer orbit
a_t = (r_c_1 + r_c_2)/2

# velocities of inner orbit
v_c_1 = sqrt(mu/r_c_1)
v_c_2 = sqrt(mu/r_c_2)

# velocity of orbit at perigee and apogee from energy equation
# (v^2)/2 + mu/r = -mu/2a

v_p_t = sqrt(2*mu /r_c_1 - mu/a_t)
v_a_t = sqrt(2*mu /r_c_2 - mu/a_t)

Δv_1 = v_p_t - v_c_1
Δv_2 = v_c_2 - v_a_t

Δv_total = Δv_1 + Δv_2

print("Δv_1",Δv_1.subs(values).evalf(),'km/s')
print("Δv_2",Δv_2.subs(values).evalf(),'km/s')
print("Δv_total",Δv_total.subs(values).evalf(),'km/s')

# transfer time
Δt = 1 / 2 * ( 2 * pi/ sqrt(mu) * (a_t)**(1.5))

print('Δt = ', (Δt / 60).subs(values).evalf(), 'minutes')
```
